```python
#Import necessary libraries
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
from PIL import Image # converting images into arrays
import matplotlib.pyplot as plt
```

```python
#import WordCloud
!pip install wordcloud
# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS
print('WordCloud and STOPWORDS are installed!')
```

```python
# download file 
#Thank you
#!wget --quiet filepath/filename.txt #wget only works with files on the web!

#print('Got the file')
```

```python
# open the file and read it into a variable
wordcloud_text = open('ThankYou.txt', 'r').read()
print(wordcloud_text)
```

```python
# stopwords are words that won't be used in the wordcloud - they were imported above
# set - the function set removes redundant stopwords.
stopwords = set(STOPWORDS)
print(stopwords)
```

```python
# instantiate a word cloud object
words_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
words_wc.generate(wordcloud_text)
```

```python
# display the word cloud
plt.imshow(words_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#Make the plot bigger
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(words_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#add words to stopwords that you don't want to use
stopwords.add('said')
stopwords.add('re')
stopwords.add('don')
```

```python
# download image to use as a mask

!wget --quiet  https://cdn.imgbin.com/2/0/1/imgbin-heart-shape-solid-fancy-heart-tattoos-dXxtEgNs6zeMMQdaFtzvkRUK3.jpg
# save mask to shape_mask
shape_mask = np.array(Image.open('imgbin-heart-shape-solid-fancy-heart-tattoos-dXxtEgNs6zeMMQdaFtzvkRUK3.jpg'))
    
print('Image downloaded and saved!')
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(shape_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Shaping the word cloud according to the mask is straightforward using word_cloud package. For simplicity, we will continue using the first 2000 words in the novel.

```python
# instantiate a word cloud object
words_wc = WordCloud(background_color='white', max_words=2000, mask=shape_mask, stopwords=stopwords,contour_width=3, contour_color='steelblue')

# generate the word cloud
words_wc.generate(wordcloud_text)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(words_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
